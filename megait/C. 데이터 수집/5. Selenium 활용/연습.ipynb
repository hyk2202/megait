```python
import chromedriver_autoinstaller
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from bs4 import BeautifulSoup
import time
from selenium.webdriver.common.by import By
chromedriver_autoinstaller.install()

driver = webdriver.Chrome()
driver.implicitly_wait(5)


driver.get("https://comic.naver.com/curation/list?type=NOVEL_ORIGIN")
time.sleep(1)
for i in range(20): # 10번 반복
    driver.execute_script("window.scrollTo(0,document.body.scrollHeight);") # 크롬 브라우져로 Javascript 코드 실행
    # 다음 컨텐츠 로딩이 완료될 때 까지 시간적 지연이 필요
    time.sleep(1)

# 브라우저에 표시되는 전체 코드를 추출 --> bs4 객체로 변환
soup = BeautifulSoup(driver.page_source)
driver.close()

```

```python
body = soup.select(".ContentList__content_list--q5KXY")[0]
content = body.select(".ContentTitle__title--e3qXt")

ranking = body.select(".rating_area")
print(len(content))
print(len(ranking))
result = []
for i in range(394):
    resultDic = {"제목": content[i].text, "별점": ranking[i].text[2:], }
    result.append(resultDic)
result
```

```python
import os 
import datetime as dt
import re
copy = result.copy()

dirname = dt.datetime.now().strftime('연습_%y%m%d_%H%M%S')

if not os.path.exists(dirname):
    os.mkdir(dirname)

for i,v in enumerate(copy):
    fname = f"{v['제목']}.txt"
    fname = re.sub('[\\/:*?<>\"\']','',fname)
    fname = f'{dirname}/{fname}'
    with open(fname,'w',encoding='utf-8') as f: # 파일만드는데 오류가 발생하면 정규표현식을 이용하여 파일명에 적히지 못하는것 제외하고 입력하면 해결
        f.write('')
    
print("fin :)")
```
