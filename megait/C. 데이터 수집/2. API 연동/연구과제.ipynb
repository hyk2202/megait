# 연구과제

## 문제 1

`https://data.hossam.kr/py/titanic.json`은 타이타닉호의 탑승자 명단을 담고 있다.

이 데이터에서 `sex` 필드는 성별을 `male`과 `female`로 표현하고 있으며 탑승객의 생존 여부는 `survived`라는 이름의 boolean 형식 값으로 저장되어 있다.

남여간의 생존자와 사망자 비율을 구하라.

```python
import requests
import json
```

### [2] 접속할 데이터의 URL

```python
url = 'https://data.hossam.kr/py/titanic.json'
```

## #02. 데이터 요청하기
### [1] 세션 생성

```python
session = requests.Session()

session.headers.update({
    "Referer" : "",
    "User_Agent" : "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/119.0.0.0 Safari/537.36"
})
```

### [2] 데이터 가져오기

```python
try: 
    r = session.get(url)
    if r.status_code != 200:
        msg = f'[{r.status_code} Error] {r.reason} 에러가 발생함'
        raise Exception(msg)
except Exception as e :# 사이트 주소가 잘못된 경우 에러발생 // 사이트주소만 올바르고 그 이후가 잘못된 경우(파일명이 잘못된 경우)는 파이썬이 감지못함
    print('접속에 실패했습니다.', e, sep = '\n')
```

### [3] 수신된 응답 결과 확인

```python
r.encoding = 'utf-8' # 읽은 후 한글이 깨지면 euc-kr로 변경
```

## #03. JSON 데이터 활용

### [1] JSON 형식의 문자열을 딕셔너리로 변환

```python
mydict = json.loads(r.text)
print(mydict)
```

```python
man_num = 0
man_survived = 0
woman_num = 0
woman_survived = 0

for i in mydict[list(mydict.keys())[0]]:
    if i['sex'] == 'male':
        man_num +=1
        if i['survived']: man_survived+=1
    else:
        woman_num +=1
        if i['survived']: woman_survived+=1
print(f'남자 생존자 비율 : {round(man_survived/man_num,2)}')
print(f'남자 사망자 비율 : {round((man_num-man_survived)/man_num,2)}')

print(f'여자 생존자 비율 : {round(woman_survived/woman_num,2)}')
print(f'여자 사망자 비율 : {round((woman_num-woman_survived)/woman_num,2)}')

```
