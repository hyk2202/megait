### [4] í•™ìŠµê³¡ì„ 

#### (1) ì „ì²´ ìƒ˜í”Œì— ëŒ€í•´ ë…ë¦½ë³€ìˆ˜ì™€ ì¢…ì†ë³€ìˆ˜ êµ¬ë¶„

```python
x = origin.drop('dist', axis=1)
y = origin['dist']
```

#### (2) í•™ìŠµê³¡ì„  ë°ì´í„° ë°˜í™˜ë°›ê¸°

í‰ê°€ì§€í‘œ(`scoring`)ì´ ì—†ì„ ê²½ìš° ê° ëª¨ë¸ë³„ë¡œ ê¸°ë³¸ í‰ê°€ì§€í‘œê°€ ì‚¬ìš©ë¨

| í‰ê°€ì§€í‘œ | ê°’ |
|---|---|
| `R2` | `r2` |
| `RMSE` | `neg_root_mean_squared_error` |
| `MSE` | `neg_mean_squared_error` |

```python
train_sizes, train_scores, test_scores = learning_curve(fit, x, y, cv=10, n_jobs=-1, train_sizes=np.linspace(0.01, 1.0, 10), scoring='neg_root_mean_squared_error')

print(train_sizes)
print(train_scores)
print(test_scores)
```

#### (3) í•™ìŠµê³¡ì„  ê·¸ë˜í”„ ê·¸ë¦¬ê¸°

```python
train_mean = -np.mean(train_scores, axis=1)
train_std = -np.std(train_scores, axis=1)
test_mean = -np.mean(test_scores, axis=1) 
test_std = -np.std(test_scores, axis=1)

plt.figure(figsize=(10, 5), dpi=200)

# í›ˆë ¨ ë°ì´í„° ìˆ˜ì— ë”°ë¥¸ í›ˆë ¨ ë°ì´í„°ì˜ score í‰ê· 
sb.lineplot(x=train_sizes, y=train_mean,  marker='o', markersize=5, label='í›ˆë ¨ë°ì´í„°', color='#ff2200')
plt.fill_between(train_sizes, train_mean + train_std, train_mean - train_std, alpha=0.15, color='#ff2200')

# ê²€ì¦ ë°ì´í„° ìˆ˜ì— ë”°ë¥¸ ê²€ì¦ ë°ì´í„°ì˜ score í‰ê· 
sb.lineplot(x=train_sizes, y=test_mean, linestyle='--', marker='s', markersize=5, label='ê²€ì¦ë°ì´í„°', color='#0066ff')
plt.fill_between(train_sizes, test_mean + test_std, test_mean - test_std, alpha=0.15, color='#0066ff')

plt.grid()
plt.xlabel('í›ˆë ¨ ì…‹íŠ¸ í¬ê¸°')
plt.ylabel('RMSE')
plt.legend()
plt.tight_layout()
plt.show()
plt.close()
```
