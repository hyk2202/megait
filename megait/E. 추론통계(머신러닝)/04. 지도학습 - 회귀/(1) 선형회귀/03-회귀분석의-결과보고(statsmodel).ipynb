### [2] ì²« ë²ˆì§¸, ì„¸ë²ˆì§¸ í‘œì˜ ë‚´ìš©

```python
my = {}

for k in [0, 2]:
    items = tbl.tables[k].data
    #print(items)

    for item in items:
        #print(item)
        n = len(item)

        for i in [0,2]:
            key = item[i].strip()[:-1]
            value = item[i+1].strip()

            if key and value:
                my[key] = value

my
```

### [3] VIFê°’ ìƒì„±

#### (1) ë‹¨ì¼ ê²°ê³¼ê°’ì„ í™•ì¸í•˜ê¸° ìœ„í•œ ì‹¤í—˜

```python
# ë°ì´í„°í”„ë ˆì„ê³¼ VIFê°’ì„ í™•ì¸í•˜ê³ ì í•˜ëŠ” ë³€ìˆ˜ì˜ ì¸ë±ìŠ¤ë¥¼ ì„¤ì •
vif = variance_inflation_factor(origin, 1)

if vif < 10:
    print("%sì˜ VIF: %f (good)" % (origin.columns[1], vif))
else:
    print("%sì˜ VIF: %f (bad)" % (origin.columns[1], vif))
```

#### (2) ëª¨ë“  ë³€ìˆ˜ì— ëŒ€í•œ ì²˜ë¦¬

```python
my['variables'] = []
name_list = list(origin.columns)

for i, v in enumerate(tbl.tables[1].data):
    #print(v)

    # ë³€ìˆ˜ì˜ ì´ë¦„
    name = v[0].strip()
    # print(name)

    if name not in name_list:
        continue

    # ë³€ìˆ˜ì˜ ì´ë¦„ ëª©ë¡ì—ì„œ í˜„ì¬ ë³€ìˆ˜ê°€ ëª‡ ë²ˆì§¸ í•­ëª©ì¸ì§€ ì°¾ê¸°
    j = name_list.index(name)

    # êµ¬í•˜ê³ ì í•˜ëŠ” ê°’
    vif = variance_inflation_factor(origin, j)

    my['variables'].append({
        "name": name,
        "coef": v[1].strip(),
        "std err": v[2].strip(),
        "t": v[3].strip(),
        "P-value": v[4].strip(),
        "Beta": 0,
        "VIF": vif,
    })

my
```

### [4] ê²°ê³¼ë³´ê³ í‘œ ë§Œë“¤ê¸°

```python
mylist = []
yname_list = []
xname_list = []

for i in my['variables']:
    yname_list.append('dist')
    xname_list.append(i['name'])

    item = {
        "B": i['coef'],
        "í‘œì¤€ì˜¤ì°¨": i['std err'],
        "Î²": i['Beta'],
        "t": i['t'],
        "ìœ ì˜í™•ë¥ " : "%s*" % i['P-value'],
        "VIF": i["VIF"]
    }

    mylist.append(item)

rdf = DataFrame(mylist,
               index=MultiIndex.from_arrays([yname_list, xname_list], names=['ì¢…ì†ë³€ìˆ˜', 'ë…ë¦½ë³€ìˆ˜']))
rdf
```
