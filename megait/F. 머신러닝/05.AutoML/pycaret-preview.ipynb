## #01. 준비작업

### [1] 패키지 가져오기

````python
# 연결된 모듈이 업데이트 되면 즉시 자동 로드함
%load_ext autoreload
%autoreload 2

from helper.util import *
from helper.plot import *
from helper.analysis import *

from pycaret.classification import ClassificationExperiment
````

### [2] 데이터 가져오기

````python
origin = my_read_excel("https://data.hossam.kr/mldata/breast-cancer-wisconsin.xlsx", index_col="id", categories=["diagnosis"], info=False)
````

### [3] 데이터 라벨링

````python
data = my_labelling(origin, 'diagnosis')
data
````

## #02. PyCaret 사용하기

### [1] 객체 초기화

분류기 객체를 생성하고 데이터와 종속변수 이름을 설정한다.

````python
s = ClassificationExperiment()
s.setup(data, target = 'diagnosis')
````

### [2] 사용할 수 있는 학습 모델의 종류 확인

````python
s.models()
````

### [3] 특정 모형 생성

`ID`값을 기준으로 특정 알고리즘에 대한 모형을 생성

하이퍼 파라미터는 기본값을 사용한다.

````python
rf = s.create_model('rf')
````

### [4] 생성된 모형에 대한 하이퍼 파라미터 튜닝

앞서 생성한 모형 객체(`rf`)를 파라미터로 전달해야 한다.

````python
tuned = s.tune_model(estimator=rf, fold=5, optimize='Accuracy')
tuned
````

### [5] 결과 시각화

````python
s.evaluate_model(estimator=tuned)
````

## #03. 앙상블

### [1] 모델 비교

`Accuracy`기준, `5`회 교차검증 수행 후 성능이 좋은 상위 5개의 모델 선정

````python
best5models = s.compare_models(sort='Accuracy', n_select=5, fold=5)
best5models
````

### [2] 선정된 모형에 대한 Voting

````python
blended = s.blend_models(estimator_list=best5models, fold=5)
blended
````

### [3] 앙상블 모형에 대한 하이퍼 파라미터 튜닝

````python
tuned = s.tune_model(estimator=blended, fold=5, optimize='Accuracy')
tuned
````

### [4] 결과 시각화

````python
s.evaluate_model(estimator=tuned)
````

### [5] 분류 결과 생성

````python
pred_holdout = s.predict_model(tuned)
pred_holdout
````
