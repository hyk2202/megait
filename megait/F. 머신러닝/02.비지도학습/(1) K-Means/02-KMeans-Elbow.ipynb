## #03. Elbow Point 구하기

### [1] 클러스터 수를 늘려가면서 inertia 값 구하기

```python
# k = 2 ~ 10까지의 inertia값을 계산
klist = list(range(2, 11))

# inertia값을 저장할 리스트
inertia = []

for k in klist:
    estimator = KMeans(n_clusters=k, random_state=0)
    estimator.fit(X=df)
    inertia.append(estimator.inertia_)
    
inertia
```

### [2] Elbow Point 구하기

| 파라미터 | 값 | 의미 |
|---|---|---|
| curve | convex | 볼록함 |
| curve | concave | 오목함 |
| direction | increasing | x가 커짐에 따라 y가 점점 커짐 |
| direction | decreasing | x가 커짐에 따라 y가 점점 작아짐 |

![img](res/knee.png)

```python
kl = KneeLocator(klist, inertia, curve="convex", direction="decreasing")
best_k = kl.elbow
print("found elbow: ", best_k)
```

## #04. 도출된 Elbow Point를 사용한 군집

### [1] 군집분석

```python
estimator = KMeans(n_clusters=best_k, random_state=0)
estimator.fit(X=df)
```

### [2] Elbow Point 시각화

```python
def my_callback(ax):
    ax.set_xlabel('Number of clusters')
    ax.set_ylabel('Inertia')
    ax.set_title('Inertia by number of clusters')
    ax.axvline(x=best_k, color='r', linestyle='--')

    ax.text(
        best_k+0.1, 1+0.1, "Best K=%d" % best_k, fontsize=18, ha="left", va="bottom", color="r"
    )

my_lineplot(df=None, xname=klist, yname=inertia, marker="o", callback=my_callback)
```
