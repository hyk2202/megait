# 메가스터디IT아카데미 산대특 데이터분석 5기 

수업 실습 자료 저장소

## 설치 패키지

| 패키지 이름 | 설명 |
|---|---|
| cx_oracle | Python을 통한 오라클 연동 기능 제공<br/>`Microsoft C++ Build Tools`를 미리 설치해야 함 <!-- <br/> : html 문법으로 줄바꿈표시 -->|
| sqlalchemy | Python을 통한 데이터베이스 연동 기능 제공(범용) |
| pandas | 분석용 데이터 구조 및 탐색적 데이터 분석 기능을 제공하는 패키지 |
| openpyxl | 엑셀 파일 처리 패키지(pandas가 의존함) |
| xlrd | 엑셀 파일 처리 패키지(pandas가 의존함) |
| requests | HTTP 요청을 처리하는 패키지 (API연동) |
| tqdm | Progressbar 구현 패키지 |
| ipywidgets | tqdm을 jupyter 상에서 동작시키는 패키지 |
| beautifulsoup4 | HTML 소스코드로부터 컨텐츠 추출 기능 제공 |
| markdownify | HTML 태그를 markdown으로 변환해 주는 패키지 |
| selenium | 크롬 브라우저를 제어하는 패키지 |
| chromedriver_autoinstaller | selenium과 크롬 브라우저를 연결해 주는 chromedirver를 자동으로 설치해 주는 패키지 |
| yfinance | 야후 파이낸스에서 크롤링하여 주가 데이터를 수집하는 패키지 |
| pytrends | 구글 트렌드 데이터를 수집하는 패키지 (`requests`, `lxml`, `pandas` 패키지에 의존한다.) |
| dotenv | 개인정보를 외부 파일에 두고 코드에서는 숨겨주는 패키지 |
| numpy | 데이터 분석의 근간이 되는 수학, 과학 연산 패키지 |

### Microsoft C++ Build Tools 다운로드

| https://data.hossam.kr/VC_redist.x64.exe

1. VC_redist.x64 : 별다른 설정 과정 없이 설치 진행
2. vs_BuildTools : 체크 항목중 왼쪽 상단의 C++ 관련 항목만 체크하고 진행
   
### 패키지 설치 구문

#### 일괄설치

```cmd
pip install --upgrade cx_oracle sqlalchemy pandas openpyxl xlrd requests tqdm ipywidgets beautifulsoup4 markdownify selenium chromedriver_autoinstaller yfinance pytrends lxml python-dotenv numpy

```
