# 피어슨 상관분석

피어슨 상관계수는 두 변수 간의 선형 상관관계를 측정

연속적인 데이터에서 사용되며, 두 변수가 서로 선형적으로 관련되어 있는 경우 유용

두 변수가 정규 분포를 따르고, 이상치(outliers)가 적은 경우에 사용 가능

값의 범위는 `-1`에서 `1`까지이며, `1`에 가까울수록 `양의 선형 상관관계`, `-1`에 가까울수록 `음의 선형 상관관계`를 의미하며, `0`에 가까울수록 `상관관계가 없음`을 의미함

## #01. 작업준비

### 패키지 가져오기

```python
import warnings
warnings.filterwarnings(action='ignore')

import numpy as np
from pandas import read_excel
from scipy.stats import pearsonr
import sys
import seaborn as sb
from matplotlib import pyplot as plt
```

```python
plt.rcParams["font.family"] = 'AppleGothic' if sys.platform == 'darwin' else 'Malgun Gothic'
plt.rcParams["font.size"] = 12
plt.rcParams["figure.figsize"] = (12, 6)
plt.rcParams['axes.unicode_minus'] = False
```

## #02 예제 (1) - 언어영역 점수와 수리영역 점수의 상관관계

주어진 데이터는 정규성과 등분산성을 충족한다고 가정한다.

### [1] 데이터 가져오기

```python
df1 = read_excel("https://data.hossam.kr/mldata/point.xlsx")
df1
```

### [2] 탐색적 데이터 분석

### [3] 공분산

공분산은 상관관계의 상승 혹은 하강하는 경향을 이해할 수는 있으나 2개 변수의 측정 단위의 크기에 따라 값이 달라지므로 절대적 정도를 파악하기에는 한계가 있다.

결과값이 양수면 정비례, 음수면 반비례

> 그래서 잘 안쓴다.

2차원 배열이 리턴되는데 이 중에서 공분산 값은 `0행1열` 혹은 `1행0열`의 값이다.

```python
c = np.cov(df1['언어영역'], df1['수리영역'])
c
```

```python
f"공분산: {c[0][1]}"
```

### [4] 상관분석

상관계수를 알아내면 된다.

#### (1) numpy를 통한 상관계수 구하기

`p-value`는 알 수 없기 때문에 분석 결과를 검정할 수 없다.

기본적으로 상관행렬을 표시한다.

```python
c = np.corrcoef(df1['언어영역'], df1['수리영역'])
c
```

상관계수를 얻기 위해서는 `0,1` 혹은 `1,0` 번째 항목을 취한다.

```python
f"상관계수: {c[0][1]}"
```

#### (2) 피어슨 상관계수와 `p-value` 구하기

```python
r, p = pearsonr(df1['언어영역'], df1['수리영역'])

f"상관계수: {r:.6f}, p-value: {p:.6f}, {'대립' if p <= 0.05 else '귀무'}가설 채택"
```

> `p-value`가 유의수준 0.05보다 작으므로 귀무가설을 기각하고 대립가설을 채택한다. 그러므로 언어영역 점수와 수리영역 점수는 상관이 있다.

### [5] 데이터프레임 객체를 통한 상관계수 구하기

이와 같이 표시되는 형태를 상관행렬이라고 한다.

`p-value` 를 확인할 수는 없다.

`scipy` 패키지의 `peqrsonr` 함수로 상ㅇ관계수와 `p-value`를 먼저 확인하고 대립가설 채택이 확정되면 시각화를 위해 이 기능을 사용한다.

```python
corr = df1.corr(method='pearson')
corr
```

#### 상관분석 결과 시각화

```python
plt.figure(figsize=(5,3))

sb.heatmap(corr,
           annot = True, #실제 값 화면에 나타내기
           cmap = 'Greens', #색상
           vmin = -1, vmax=1 , #컬러차트 영역 -1 ~ +1
           linewidths=.5 #구분선
)

plt.show()
plt.close()
```

## #03 예제 (2) - mtcars 데이터를 통한 실린더수와 연비의 상관관계 분석

R 패키지인 ggplot2에 내장된 샘플 데이터로 자동차 32종의 11개 속성에 대한 정보를 담고 있는 데이터

| 필드 | 설명 |
|---|---|
| mpg | 연료 효율성 (마일 당 갤런) |
| cyl | 실린더 수 |
| disp | 배기량 (cubic inches) |
| hp | 마력 (horsepower) |
| drat | 후륜축 비율 (rear axle ratio) |
| wt | 차량 무게 (1000 파운드) |
| qsec | 1/4 마일 달성 시간 (초) |
| vs | V/S (V 엔진 형태인지, S는 직렬 엔진 형태인지를 나타냄) |
| am | 변속기 형태 (0: 자동, 1: 수동) |
| gear | 전진 기어 수 (전송 기어) |
| carb | 기화기 개수 (carburetors) |

### [1] 데이터 가져오기

```python
df2 = read_excel("https://data.hossam.kr/mldata/mtcars.xlsx", index_col='name')
df2
```

### [2] 배기량에 따른 연비에 대한 산점도 그래프

```python
plt.figure(figsize = (10, 5))

sb.regplot(data=df2, x='disp', y='mpg')
plt.grid()
plt.show()
plt.close()
```

> 배기량이 증가함에 따라 연비가 낮아지는 경향이 있는것으로 보임


### [3] 상관분석

```python
s, p = pearsonr(df2['disp'],df2['mpg'])
f"상관계수: {s:.3f}, p-value: {p:.3f}, {'대립' if p <= 0.05 else '귀무'}가설 채택"
```

### [4] 분석결과 시각화


```python
plt.figure(figsize= (5,3), dpi=200)

corr = df2.filter(['disp', 'mpg']).corr(method='pearson')

sb.heatmap(corr,
           annot = True, #실제 값 화면에 나타내기
           cmap = 'Blues', #색상
           vmin = -1, vmax=1 , #컬러차트 영역 -1 ~ +1
           linewidths=.5 #구분선
)

plt.show()
plt.close()
```

> 32종의 자동차를 대상으로 배기량에 따른 연비를 조사한 결과 배기량과 연비는 강한 음의 상관관계를 갖는 것으로 타나났다. `p-value`가 유의수준 0.05보다 작으므로 이는 통계적으로 유의하다.
